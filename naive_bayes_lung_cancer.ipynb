```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/lung_cancer_examples.csv")
df
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df["Name"] = lb.fit_transform(df["Name"])
df["Name"]
```

```python
df["Surname"]=lb.fit_transform(df["Surname"])
df["Surname"]
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train
```

```python
x_test
```

```python
#Noramalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.naive_bayes import BernoulliNB
base=BernoulliNB() #creating variable
base.fit(x_train,y_train) #the data teaching to machine
y_pred=base.predict(x_test) #the machine predicts value according to the model
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_pred,y_test)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
